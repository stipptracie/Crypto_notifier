# Determine threshhold for "big swing" in each coin


```python
# Import required libraries and dependencies

import pandas as pd
import os
import json
import requests
import numpy as np
from pycoingecko import CoinGeckoAPI
```

```python
# Set Variable for coingecko API
cg = CoinGeckoAPI()
```

# Bitcoin (BTC)

```python
# Call Bitcoin data from CoinGecko and create dataframe.  Date range from 01/01/2021 - 07/01/2022.

historical_btc = cg.get_coin_market_chart_range_by_id(id='bitcoin',vs_currency='usd',from_timestamp='1609515440',to_timestamp='1656689840')
btc_df = pd.DataFrame(historical_btc['prices'], columns=['unix_timestamp', 'price']).set_index('unix_timestamp')
btc_df.head()
```

```python
#calculate pct_change, and turn into an absolute value
btc_change_df = pd.DataFrame(btc_df['price'].pct_change().dropna()).rename(columns={'price':'absolute_change'})
btc_change_df = np.absolute(btc_change_df)
btc_change_df
```

```python
#calculate mean of pct_change
btc_threshold = btc_change_df.mean()
print(btc_threshold)
```
