```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
import time
import datetime
```

```python
def unix_time(year, month, day, hour, second):
    date_time = datetime.datetime(year, month, day, hour, second)
    return time.mktime(date_time.timetuple())
def human_time(unix_time):
    return datetime.datetime.fromtimestamp(unix_time)
```

```python
start_time = unix_time(2021, 9, 16, 0, 0)
end_time = unix_time(2022, 9, 16, 0, 0)
print('start_time:', start_time)
print('end_time:', end_time)  
```

```python

cg = CoinGeckoAPI()
btc = cg.get_coin_market_chart_range_by_id(id='bitcoin',vs_currency='usd',from_timestamp=start_time, to_timestamp=end_time)
btc_df = pd.DataFrame(btc)


btc_prices_1_year = pd.DataFrame(btc_df['prices'].tolist())
btc_prices_1_year.columns =['Date','Price']
btc_prices_1_year['Date'] = pd.to_datetime(btc_prices_1_year['Date'], unit='ms')


btc_prices_1_year
```

```python
eth = cg.get_coin_market_chart_range_by_id(id='ethereum',vs_currency='usd',from_timestamp=start_time, to_timestamp=end_time)
eth_df = pd.DataFrame(eth)


eth_prices_1_year = pd.DataFrame(eth_df['prices'].tolist())
eth_prices_1_year.columns =['Date','Price']
eth_prices_1_year['Date'] = pd.to_datetime(eth_prices_1_year['Date'], unit='ms')


eth_prices_1_year
```
